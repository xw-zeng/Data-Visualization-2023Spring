## Create a Graph

Create an empty graph with no nodes and no edges.

`````python
import networkx as nx
`````

`````python
G = nx.Graph()
`````

By definition, a Graph is a collection of nodes (vertices) along with identified pairs of nodes (called edges, links, etc).

## Nodes

To get started though weâ€™ll look at simple manipulations. You can add one node at a time,

`````python
G.add_node(1)
`````

or add nodes from any iterable container, such as a list

`````python
G.add_nodes_from([1, 2, 3, 4])
`````

## Edges

G can also be grown by adding one edge at a time,

`````python
G.add_edge(1, 2)
`````

by adding a list of edges,

`````python
G.add_edges_from([(1, 2), (1, 3), (2, 3), (2, 4)])
`````

## Drawing graphs

`````python
print("# of nodes: {}, # of edges: {}".format(G.number_of_nodes(), G.number_of_edges()))
`````

`````python
nx.draw(G, with_labels=True, node_size=500)
`````

## Get the adjacency matrix 

`````python
As = nx.adjacency_matrix(G)
`````

`````python
print(As)
`````

`````python
A = As.todense()
`````

`````python
print(A)
`````

## Construct a graph from an adjacency matrix 

`````python
import numpy as np

A = np.array([[0, 1, 1], [1, 0, 1], [1, 1 ,0]])
`````

`````python
G = nx.from_numpy_array(A)
`````

`````python
nx.draw(G, with_labels=True, node_size=500)
`````
